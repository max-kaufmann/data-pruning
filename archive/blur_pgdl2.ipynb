```python
def normalize_blur_l2(blur):
    '''
    a x b blur (for only one image--not batched)
    '''
    norm = torch.norm(blur.view(-1), p=2)
    return blur / norm

def blur_creator(vars, image):
    '''
    Creates some noise based on some variable. 
    Creation of noise should be batched and enable gradients 
    to flow through it.
    '''
    
    x = image.clone()
    x = torchvision.transforms.functional.gaussian_blur(x, kernel_size=51, sigma=15)
    x = x.squeeze(0)
    
    blur = vars[0] # Between 0 and 1, represents the scale of the blur?
    blur = blur.unsqueeze(0)
    blur = torchvision.transforms.functional.gaussian_blur(blur, kernel_size=51, sigma=15)
    blur = blur.squeeze(0)
    
    blurred = x * blur + image * (1-blur)
    blurred = blurred.clamp(0, 1)
    return blurred

def tensor_clamp_blur_l2(x, center, radius):
    """clamp of x into l2 ball around center of given radius. not batched."""
    x = x.data
    diff = x - center
    diff_norm = torch.norm(diff.view(-1), p=2)
    project_select = diff_norm > radius
    if project_select:
        print(diff_norm, radius)
        diff_norm = diff_norm.unsqueeze(-1).unsqueeze(-1)
        new_x = x
        new_x = (center + (diff / diff_norm) * radius)
        
        new_diff = new_x - center
        new_diff_norm = torch.norm(new_diff.view(-1), p=2)
        assert new_diff_norm <= radius + 0.01
        return new_x
    else:
        return x
```
